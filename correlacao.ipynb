```python
# library import

import pandas as pd
import seaborn as sn
```

```python
# leitura dos dados

data = pd.read_csv("https://raw.githubusercontent.com/wellalb/Data-Science/main/forestfires.csv")
print(data)
```

```python
# an√°lise de correla√ß√£o

correlation = data.corr()
```

```python
# plot da matriz de correla√ß√£o

plot = sn.heatmap(correlation, annot = True, fmt=".1f", linewidths=.6)
plot
```
